```python
import pandas as pd
import datetime as dt
now = dt.datetime.now()
print(now)
print('\n*****\n')
```

```python
dr = pd.date_range(now,freq='Y',periods=5)

print(dr)
```

```python
future = now + dt.timedelta(100)

dr = pd.date_range(now,future,freq='w')

print(dr)
```

```python
df = pd.DataFrame(dr,columns=['dtg'])

df['date'] = df['dtg'].dt.date

df['time'] = df.dtg.dt.time

df['month'] = df.dtg.dt.to_period('M')

df['year'] = df.dtg.dt.to_period('Y')

df
```

```python
df['dtg_str'] = df.dtg.astype(str)

df['dtg2'] = pd.to_datetime(df.dtg_str)



df['dtg_str_cust'] = '2021ff03ff01'

df['dtg_str_frmt']=pd.to_datetime(df.dtg_str_cust,format="%Yff%mff%d",errors='coerce')



df
```
