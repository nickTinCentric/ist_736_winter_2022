```python
#create state data
state_data = {'State':['Alabama','Alaska','Arizona','Arkansas'],'PostCode':['AL','AK','AZ','AR'],'Area':['52,423','656,424','*','53,182'],'Pop':['4,040,587','550,043','3,665,228','2,350,750']}

```

```python
#create data data frame
import pandas as pd

stdf = pd.DataFrame(state_data, columns=['State', 'PostCode','Area','Pop'])

stdf
```

```python
#index table by state

stdf = stdf.set_index('State')

stdf
```

```python
#4. Replace the '*' with '0'.

stdf = stdf.replace('*', '0')

stdf
```

```python
#5. Define a function to replace ',' with ''.

def replace_comma(xdf):
    return xdf.replace(',', '')
```

```python
#6. Use this function to replace the commas in 'Area' and 'Pop'.

stdf['Area'] = stdf['Area'].map(replace_comma)
stdf['Pop'] = stdf['Pop'].map(replace_comma)

stdf
```
