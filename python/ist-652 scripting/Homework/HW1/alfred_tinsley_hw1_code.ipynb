```python
## build regseson subset, at this point i want to see goalie stats during regular season starting in the year
columns =["playerID", 'stint', 'tmID', 'GP', 'Min', 'W', 'L', 'T/OL', 'ENG', 'SHO', 'GA', 'SA']
regseason = df.loc[(df['year']>=2000) & (df['lgID']== 'NHL'), columns]
regseason
```

```python
## i want to see saves, so it would be SA = shots against - GA Goals against to get the amount of saves a goalie made
regseason['saves'] = regseason['SA'] - regseason['GA']
```

```python
## organize regseason data, group by year, team and return top winning goalies
# df['INCOME'].groupby([df['homeowner dummy'],df['AVGGIFT'],df['gender dummy']]).mean()

by_player = regseason.groupby([regseason["playerID"],regseason["tmID"]]).mean()

by_player
```
